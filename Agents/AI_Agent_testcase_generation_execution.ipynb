```python
import autogen

config_list = autogen.config_list_from_json(
    env_or_file="/content/drive/MyDrive/LLM/llm.json",
    filter_dict={
        "model": ["gpt-4"],
    },
)
print(config_list)
```

```python
llm_config={
    "timeout": 600,
    "cache_seed": 42,
    "config_list": config_list,
    "temperature": 0,
}
```

```python
# create an AssistantAgent instance named "assistant"
assistant = autogen.AssistantAgent(
    name="assistant",
    llm_config=llm_config,
)
# create a UserProxyAgent instance named "user_proxy"
user_proxy = autogen.UserProxyAgent(
    name="user_proxy",
    human_input_mode="TERMINATE",
    max_consecutive_auto_reply=10,
    is_termination_msg=lambda x: x.get("content", "").rstrip().endswith("TERMINATE"),
    code_execution_config={"work_dir": "web"},
    llm_config=llm_config,
    system_message="""Reply TERMINATE if the task has been solved at full satisfaction.
Otherwise, reply CONTINUE, or the reason why the task is not solved yet."""
)
```

```python
# the assistant receives a message from the user, which contains the task description
user_proxy.initiate_chat(
    assistant,
    message="""
Task: Test Case Generation for REQ_SPE.pdf

File: /content/drive/MyDrive/Data/REQ_SPE.pdf

Objective:
Generate all possible test cases related to the specified PDF file.

Instructions:
Identify diverse scenarios in the PDF content.
Create test cases covering different aspects (e.g., edge cases, valid inputs).
Save generated test cases in a file.

Submission:
save the file containing the generated test cases.
""",
)
```

```python
# the assistant receives a message from the user, which contains the task description
user_proxy.initiate_chat(
    assistant,
    message="""
Task: Program Development for REQ_SPE.pdf Requirements

File: /content/drive/MyDrive/Data/REQ_SPE.pdf

Objective:
Develop a program based on the requirements outlined in the specified PDF file.

Instructions:
Extract and comprehend the key requirements from the PDF content.
Write a program addressing the identified requirements.
Ensure the program adheres to specified functionalities.

Submission:
Save the program file with calculator.py .
""",
)
```

```python
# the assistant receives a message from the user, which contains the task description
user_proxy.initiate_chat(
    assistant,
    message="""
Files:
Program: /content/web/calculator.py
Test Cases: /content/web/generate_test_cases.py

Objective:
Run test cases from write_test_cases.py on calculator.py.
Check pass/fail status for each test case.
Provide a report summarizing results.

Instructions:
Use a testing framework or method for automation.
Note errors or unexpected behavior.
Include details for failed test cases.

Submission:
If all test cases are not passed modified those test cases so that all the test cases must passed.
And Modified the program as well according to the modified test case.
""",
)
```

```python
# the assistant receives a message from the user, which contains the task description
user_proxy.initiate_chat(
    assistant,
    message="""
Task: Test Case Modification and Execution

Files:
Program File: /content/web/calculator.py
Original Test Cases File: /content/web/test_runner.py

Objective:
Modify the test cases in write_test_cases.py to align with the program in calculator.py.
Execute the modified test cases on the calculator program.
Verify the pass/fail status of each test case.

Instructions:
Review the program (calculator.py) to understand its functionalities.
Adjust the existing test cases in write_test_cases.py accordingly.
Run the modified test cases on the program.
Check and document the results (pass/fail) for each test case.

Submission:
Save the modified test cases file and a report on the test results.
""",
)
```

```python
# followup of the previous question
user_proxy.send(
    recipient=assistant,
    message="""
    Task: Validation Check for Test Cases

Objective:
Ensure the code in /content/web/calculator.py passes all test cases.
Validate that the output of each test case is correct.

Instructions:
Execute the test cases from /content/web/test_runner.py on calculator.py.
Verify the correctness of the output for each test case.
Confirm that the program has passed all test cases successfully.

Submission:
Save a report confirming the validation status and highlight any discrepancies.
    """,
)
```
